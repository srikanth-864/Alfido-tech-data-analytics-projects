# KPI ANALYSIS

```python
print( "Total Events:", len(df) )
```

```python
#Count Event Types

df["event"].value_counts()
```

# Traffic Events Visualization

```python
df["event"].value_counts().plot( kind="bar" )
plt.title( "Traffic Event Distribution" ) 
plt.xlabel( "Event" ) 
plt.ylabel( "Count" ) 
plt.show()
```

# Top Countries visualization

```python
df[ "country" ].value_counts().head( 10 ).plot( kind="bar" )
plt.title( "Top Countries" )
plt.show()
```

# Top Artists Visualization

```python
df[ "artist" ].value_counts().head( 10 ).plot( kind="bar" ) 
plt.title( "Top Artists" )
plt.show()
```

# Daily Traffic Trend visualization

```python
daily=df.groupby( "date" ).size() 
daily.plot( kind="line" )
plt.title( "Daily Traffic" )
plt.show()
```
